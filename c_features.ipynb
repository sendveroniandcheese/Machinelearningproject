```python
from sklearn import tree
import pandas as pd
```

```python
df = pd.read_csv('nba_df.csv')
df
```

```python
grouped = df.groupby('Pos')
```

```python
c_df = grouped.get_group('C')

c_df
```

```python
target = df["All_Star"]
target_names = ["All Star", "Loser"]

data = df.drop(["All_Star", 'Player', 'Pos', 'Age', 'Tm' ], axis=1)
data
```

```python
feature_names = data.columns  
#feature_names
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=43)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=250)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
c_features = sorted(zip(rf.feature_importances_, feature_names), reverse=True)
c_features
```
