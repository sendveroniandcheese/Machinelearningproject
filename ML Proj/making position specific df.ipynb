```python
from sklearn import tree
import pandas as pd
```

```python
df = pd.read_csv('nba_df.csv')
df
```

```python
grouped = df.groupby('Pos')
```

```python
pg_df = grouped.get_group('PG')

pg_df
```

```python
target = df["All_Star"]
target_names = ["All Star", "Loser"]
```

```python
data = df.drop(["All_Star", 'Player', 'Pos', 'Age', 'Tm' ], axis=1)
data
```

```python
feature_names = data.columns  
#feature_names
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=43)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=250)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
#pg specific all star criteria 
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
