```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = "clean20152020.csv"

# Import the CSV into a pandas DataFrame
nba_df = pd.read_csv(csv_path, low_memory=False)
nba_df
```

```python
nba_df.fillna('no', inplace=True)
```

```python
nba_df.head(20)
```

```python
result_df = nba_df.sort_values('year')

result_df.head(44)

sorted_df = result_df.sort_values('year')
cleaner_nba = sorted_df.drop_duplicates()


cleaner_nba.to_csv('nba_df.csv', index=False)
```

```python
cleaner_nba.rename(columns = {'All Star':'All_Star'}, inplace = True)
```

```python
cleaner_nba.All_Star.replace(('Yes', 'no'), (1, 0), inplace=True)

cleaner_nba
```
